```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('C:/Users/ADM/OneDrive/Documentos/IEEE CIS UnB/DataSets/winequality.csv')
print(len(dataset))
print(dataset.head())
```

```python
dataset.head()
```

```python
dataset.isnull().any()
dataset.isnull().sum()
```

```python
#Separar o Dataset
x = dataset.iloc[:, 0:13]
y = dataset.iloc[:, 13]
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 0, test_size= 0.2)
```

```python
#Dimensionar os recursos
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

KNN

```python
# Iniciando o KNN
classifier = KNeighborsClassifier(n_neighbors = 71, p = 2, metric = 'euclidean')
```

```python
classifier.fit(x_train,y_train)
```

```python
# Prever o resultado dos conjuntos de teste
y_pred = classifier.predict(x_test)
y_pred
```

```python
# Avaliar o modelo
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(f1_score(y_test, y_pred))
```

```python
# Acurácia dos dados
print(accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('C:/Users/ADM/OneDrive/Documentos/IEEE CIS UnB/DataSets/winequality.csv')
```

```python
df.head()
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
df['wine_is_red'].unique()
```

```python
len(df)
```

```python
vinho_outro = df[df['wine_is_red'] == 0]
vinho_tinto = df[df['wine_is_red'] == 1]
```

```python
vinho_outro_downsampled = resample(vinho_outro,replace=False,n_samples=1000,random_state=42)
len(vinho_outro_downsampled)
```

```python
vinho_tinto_downsampled = resample(vinho_tinto,replace=False,n_samples=1000,random_state=42)
len(vinho_tinto_downsampled)
```

```python
df_downsample = pd.concat([vinho_outro_downsampled,vinho_tinto_downsampled])
len(df_downsample)
```

```python
x = df_downsample.drop('wine_is_red',axis=1).copy()
x.head()
```

```python
y = df_downsample['wine_is_red'].copy()
y.head()
```

```python
df_downsample['quality'].unique()
```

```python
x_encoded = pd.get_dummies(x,columns=['quality'])
x_encoded.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_encoded,y,random_state=42)
x_train_scaled = scale(x_train)
x_test_scaled = scale(x_test)
```

```python
# SVM Preliminar
clf_svm = SVC(random_state=42)
clf_svm.fit(x_train_scaled,y_train)
```

```python
plot_confusion_matrix(clf_svm,x_test_scaled,y_test,values_format='d',display_labels=['Outro Vinho','Vinho Tinto'])
```

```python
#SVM com validação cruzada
param_grid = [{'C':[0.5,1,10,100],'gamma':['scale',1,0.1,0.001,0.0001],'kernel':['rbf']},]
```

```python
optimal_params = GridSearchCV(SVC(),param_grid,cv=5,scoring='accuracy',verbose=0)
```

```python
optimal_params.fit(x_train_scaled,y_train)
print(optimal_params.best_params_)
```

```python
clf_svm = SVC(random_state=42,C=100,gamma=0.001)
clf_svm.fit(x_train_scaled,y_train)
```

```python
plot_confusion_matrix(clf_svm,x_test_scaled,y_test,values_format='d',display_labels=['Outro Vinho','Vinho Tinto'])
```

```python
len(df_downsample.columns)
```

```python
pca = PCA()
x_train_pca = pca.fit_transform(x_train_scaled)
```

```python
per_var = np.round(pca.explained_variance_ratio_*100,decimals=1)
labels = [str(x) for x in range(1,len(per_var)+1)]
```

```python
plt.bar(x=range(1,len(per_var)+1),height=per_var)
plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.ylabel('Porcentagem de Variância Explicada')
plt.xlabel('Principais componentes')
plt.title('Scree Plot')
plt.show()
```

```python
train_pc1_coords = x_train_pca[:,0]
train_pc2_coords = x_train_pca[:,1]
```

```python
pca_train_scaled = scale(np.column_stack((train_pc1_coords,train_pc2_coords)))
```

```python
param_grid = [{'C': [1,10,100,1000],'gamma':['scale',1,0.1,0.001,0.0001],'kernel':['rbf']},]
```

```python
optimal_params = GridSearchCV(SVC(),param_grid,cv=5,scoring='accuracy',verbose=0)
optimal_params.fit(pca_train_scaled,y_train)
print(optimal_params.best_params_)
```

```python
clf_svm = SVC(random_state=42, C=1000, gamma=0.001)
clf_svm.fit(pca_train_scaled,y_train)
```

```python
x_test_pca = pca.transform(x_train_scaled)
test_pc1_coords = x_test_pca[:,0]
test_pc2_coords = x_test_pca[:,1]
```

```python
x_min = test_pc1_coords.min() - 1
x_max = test_pc2_coords.max() + 1

y_min = test_pc2_coords.min() - 1
y_max = test_pc2_coords.max() + 1
```

```python
xx,yy = np.meshgrid(np.arange(start=x_min,stop=x_max,step=0.1),np.arange(start=y_min,stop=y_max,step=0.1))
```

```python
z = clf_svm.predict(np.column_stack((xx.ravel(),yy.ravel())))
z = z.reshape(xx.shape)
```

```python
fig , ax = plt.subplots(figsize=(10,10))
ax.contourf(xx,yy,z,alpha=0.1)
cmap = colors.ListedColormap(['#e41a1c','#4daf4a'])
scatter = ax.scatter(test_pc1_coords,test_pc2_coords,c=y_train,cmap=cmap,s=100,edgecolors='k',alpha=0.7)
legend = ax.legend(scatter.legend_elements()[0],scatter.legend_elements()[1],loc='upper right')
legend.get_texts()[0].set_text('Outro Vinho')
legend.get_texts()[1].set_text('Vinho Tinto')

ax.set_ylabel('PC2')
ax.set_xlabel('PC1')
ax.set_title('Superfície de decisão usando as características transformadas/projetadas do PCA')
plt.show()
```

Oversample e Undersample

```python
df = pd.read_csv('C:/Users/ADM/OneDrive/Documentos/IEEE CIS UnB/DataSets/winequality.csv')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head()
```

```python
def transformar_wine(valor):
    if valor == 0:
        return 'outro'
    else:
        return 'tinto'
```

```python
df['tinto/outro'] = df['wine_is_red'].map(transformar_wine)
```

```python
df
```

```python
x = df.drop(['tinto/outro'],axis=1)
y = df['tinto/outro']
```

Class distribution

```python
y.value_counts()
```

```python
y.value_counts().plot.pie(autopct='%.2f')
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
#Undersampling
rus = RandomUnderSampler(sampling_strategy=1)
x_res, y_res = rus.fit_resample(x, y)

ax = y_res.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title("Under-sampling")
```

```python
y_res.value_counts()
```

```python
#Overampling
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(sampling_strategy="not majority")
x_res, y_res = ros.fit_resample(x, y)

ax = y_res.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title("Over-sampling")
```
