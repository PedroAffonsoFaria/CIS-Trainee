AtivaÃ§Ã£o do Percepton

```python
# FunÃ§Ã£o de ativaÃ§Ã£o sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivada da funÃ§Ã£o sigmoide
def sigmoid_derivada(x):
    return x * (1 - x)
```

```python
df
```

```python
# Separar os dados em features (X) e labels (y)
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values.reshape(-1, 1)
```

```python
# Dividir os dados em conjuntos de treinamento e teste
split_proporcao = 0.8
split_indice = int(split_proporcao * len(X))
X_train, X_test = X[:split_indice], X[split_indice:]
y_train, y_test = y[:split_indice], y[split_indice:]
```

```python
# ConfiguraÃ§Ã£o do perceptron
input_tamanho = X.shape[1]
hidden_tamanho = 4  # Tamanho da camada oculta
taxa_de_aprendizagem = 0.1
epocas = 10000

# InicializaÃ§Ã£o aleatÃ³ria dos pesos e bias para a camada oculta e a camada de saÃ­da
np.random.seed(42)
pesos_ocultos = np.random.uniform(size=(input_tamanho, hidden_tamanho))
bias_ocultos = np.random.uniform(size=(1, hidden_tamanho))
pesos_output = np.random.uniform(size=(hidden_tamanho, 1))
bias_output = np.random.uniform(size=(1, 1))
print('Pesos inicializados:',pesos_output)
print('Bias:',bias_output)
```

```python
# Treinamento
for epoca in range(epocas):
    # Forward pass na camada oculta
    hidden_input = np.dot(X_train, pesos_ocultos) + bias_ocultos
    hidden_output = sigmoid(hidden_input)
    
    # Forward pass na camada de saÃ­da
    output = sigmoid(np.dot(hidden_output, pesos_output) + bias_output)

    # CÃ¡lculo do erro
    erro = y_train - output
    
    # Backpropagation
    d_output = erro * sigmoid_derivada(output)
    d_hidden = np.dot(d_output, pesos_output.T) * sigmoid_derivada(hidden_output)
    
    # AtualizaÃ§Ã£o dos pesos e bias
    pesos_output += np.dot(hidden_output.T, d_output) * taxa_de_aprendizagem
    bias_output += np.sum(d_output, axis=0, keepdims=True) * taxa_de_aprendizagem
    pesos_ocultos += np.dot(X_train.T, d_hidden) * taxa_de_aprendizagem
    bias_ocultos += np.sum(d_hidden, axis=0, keepdims=True) * taxa_de_aprendizagem
    
print(pesos_output)
```

```python
# Calcular as previsÃµes para o conjunto de treinamento
hidden_input_train = np.dot(X_train, pesos_ocultos) + bias_ocultos
hidden_output_train = sigmoid(hidden_input_train)
output_train = sigmoid(np.dot(hidden_output_train, pesos_output) + bias_output)
train_predictions = np.round(output_train).astype(int)
print(train_predictions)
```

```python
# Calcular as previsÃµes para o conjunto de teste
hidden_input_test = np.dot(X_test, pesos_ocultos) + bias_ocultos
hidden_output_test = sigmoid(hidden_input_test)
output_test = sigmoid(np.dot(hidden_output_test, pesos_output) + bias_output)
test_predictions = np.round(output_test).astype(int)
```

```python
# Medir a acurÃ¡cia do treinamento sem usar mÃ©dia
train_correct_predictions = np.sum(train_predictions == y_train)
train_accuracy = train_correct_predictions / len(y_train) * 100
print(f"AcurÃ¡cia do Treinamento: {train_accuracy:.2f}%")

# Medir a acurÃ¡cia do teste sem usar mÃ©dia
test_correct_predictions = np.sum(test_predictions == y_test)
test_accuracy = test_correct_predictions / len(y_test) * 100
print(f"AcurÃ¡cia do Teste: {test_accuracy:.2f}%")
```

**ValidaÃ§Ã£o Cruzada para verificar Overfitting**

```python
from sklearn.model_selection import KFold

# NÃºmero de partes para a validaÃ§Ã£o cruzada
n_folds = 5

# InicializaÃ§Ã£o do validador cruzado
kf = KFold(n_splits=n_folds)

# Listas para armazenar as acurÃ¡cias de treinamento e teste para cada fold
acuracias_treino = []
acuracias_teste = []
```

```python
# ValidaÃ§Ã£o cruzada
for train_indice, test_indice in kf.split(X):
    X_train, X_test = X[train_indice], X[test_indice]
    y_train, y_test = y[train_indice], y[test_indice]

    # Restante do treinamento e teste (como no seu cÃ³digo original)

    # Calcular as previsÃµes para o conjunto de treinamento
    hidden_input_train = np.dot(X_train, pesos_ocultos) + bias_ocultos
    hidden_output_train = sigmoid(hidden_input_train)
    output_treino = sigmoid(np.dot(hidden_output_train, pesos_output) + bias_output)
    treino_predictions = np.round(output_treino).astype(int)

    # Calcular as previsÃµes para o conjunto de teste
    hidden_input_test = np.dot(X_test, pesos_ocultos) + bias_ocultos
    hidden_output_test = sigmoid(hidden_input_test)
    output_teste = sigmoid(np.dot(hidden_output_test, pesos_output) + bias_output)
    teste_predictions = np.round(output_teste).astype(int)

    # Medir a acurÃ¡cia do treinamento sem usar mÃ©dia
    acuracia_treino =  np.sum(treino_predictions == y_train) / len(y_train) * 100
    acuracias_treino.append(acuracia_treino)

    # Medir a acurÃ¡cia do teste sem usar mÃ©dia
    acuracia_teste = np.sum(teste_predictions == y_test) / len(y_test) * 100
    acuracias_teste.append(acuracia_teste)

# MÃ©dias das acurÃ¡cias para todos os folds
med_acuracia_treino = np.mean(acuracias_treino)
med_acuracia_teste = np.mean(acuracias_teste)

print(f"AcurÃ¡cia MÃ©dia do Treinamento: {med_acuracia_treino:.2f}%")
print(f"AcurÃ¡cia MÃ©dia do Teste: {med_acuracia_teste:.2f}%")
```
